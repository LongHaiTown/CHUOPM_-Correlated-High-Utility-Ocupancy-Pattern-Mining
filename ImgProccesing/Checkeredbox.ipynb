# Create a chess table using numpy and openCV

```python

import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
windows_WIDTH_HEIGTH = 1280;
tile_WIDTH_HEIGHT = (int)(windows_WIDTH_HEIGTH/8);
```

```python
blacka = np.zeros((tile_WIDTH_HEIGHT,tile_WIDTH_HEIGHT,3))
whitenoise=  np.ones((tile_WIDTH_HEIGHT,tile_WIDTH_HEIGHT, 3))

plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
plt.imshow(whitenoise,cmap='gray')

plt.subplot(2,2,2)
plt.imshow(blacka,cmap='gray')
plt.axis("off")

```

```python
board = np.ones((windows_WIDTH_HEIGTH, windows_WIDTH_HEIGTH, 3), dtype=np.uint8) *255
```

```python
array_2d = np.zeros((8, 8))
for number in range(0,len(array_2d)):
    for number1 in range(0,len(array_2d)):
        if (number%2== 0):
            if(number1%2 ==0):
                array_2d[number][number1] =1;
        else: 
            if(number1%2 !=0):
                array_2d[number][number1] =1;
array_2d
```

```python
current_X= 0
current_Y= 0
for number in range(0,8):
    current_X = 0  
    for number1 in range(0,8):
        if (array_2d[number][number1] == 1):
            board[current_X:(current_X + tile_WIDTH_HEIGHT), current_Y:(current_Y + tile_WIDTH_HEIGHT)] = blacka
        current_X += tile_WIDTH_HEIGHT

    current_Y += tile_WIDTH_HEIGHT
    
print(current_Y)
```

```python
plt.imshow(board,cmap='gray')
plt.axis("off")
```
