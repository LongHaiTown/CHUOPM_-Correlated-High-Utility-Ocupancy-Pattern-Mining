```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students = pd.read_csv('students.csv')
students.head(10)
students.describe(include='all')
```

```python
mean_math_grade = students.math_grade.mean()
median_math_grade = students.math_grade.median()
mode_math_grade = students.math_grade.mode()[0]
math_grade_range = students.math_grade.max() - students.math_grade.min()
math_grade_std = students.math_grade.std()

print(mean_math_grade)
print(median_math_grade)
print(mode_math_grade)
print(math_grade_range)
print(math_grade_std)
```

### Visualizing

```python
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.hist(students.math_grade)
plt.title("Histogram")


plt.subplot(122)
sns.boxplot(students.math_grade)
plt.title("Box plot")

plt.show()
plt.clf()
```

```python
mother_job = students['Mjob'].value_counts(normalize = True)
print(mother_job)
```

```python

plt.figure(figsize=(16,9))
plt.subplot(121)
sns.countplot(x='Mjob',data= students)
plt.title("Histogram")


plt.subplot(122)
students.Mjob.value_counts().plot.pie()
plt.title("Box plot")
plt.ylabel("")

plt.show()
plt.clf()
```
